## Pré-traitement des données

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

SEED = 42

X = df.drop(columns=['fetal_health'])
y = df['fetal_health']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=SEED)
```

## Modèles

Dans cette section nous présenterons deux modèles que nous avons entraînés pour ce travail. Un arbre de décision, très interprétable et une random forest, un peu moins interprétable mais potentiellement plus performante.

### Arbre de décision

```python
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

def show_confusion_matrix(y_true, y_pred, show_report=True):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=['normal', 'suspect', 'pathological'],
                yticklabels=['normal', 'suspect', 'pathological'])
    plt.xlabel('Prédictions')
    plt.ylabel('Vérités')
    plt.title('Matrice de confusion')
    plt.show()
    if show_report:
        print(classification_report(y_true, y_pred, target_names=['normal', 'suspect', 'pathological']))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

dt = DecisionTreeRegressor(random_state=SEED)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

show_confusion_matrix(y_test, y_pred_dt)

features = list(df.columns.values)

plt.figure(figsize=(30,10))
tree.plot_tree(dt, max_depth=2, feature_names=features, filled=True, fontsize=10, impurity=False)
```

### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
        n_estimators=400,
        max_depth=None,
        min_samples_leaf=2,
        class_weight="balanced",
        random_state=SEED,
        n_jobs=-1)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

show_confusion_matrix(y_test, y_pred_rf)
```

### Comparaison des résultats

| Modèle                 | Precision | Recall | F1-score |
|------------------------|-----------|--------|----------|
| DecisionTreeClassifier | 0.89      | 0.90   | 0.89     |
| RandomForestClassifier | 0.93      | 0.93   | 0.93     |

On observe que la RandomForest est légèrement meilleure.

## Explication post-hoc

Nous utiliserons l'arbre de décision créé ci-dessus pour illustrer l'utilisation des méthodes LIME et SHAP.

### Explication globale

```python
def show_feature_importances(re_importances):
    # Create a figure with one subplot
    fig, ax = plt.subplots(figsize=(10, 5))

    # Plot the second set of data
    sorted_idx = re_importances.argsort()[::-1]
    sorted_importances = re_importances[sorted_idx][::-1]
    sorted_features = X.columns[sorted_idx][::-1]

    ax.barh(sorted_features, sorted_importances)
    ax.set_xlabel('Importance')
    ax.set_ylabel('Feature')
    ax.set_title('Classifier Feature Importances')
    ax.yaxis.set_ticks(sorted_features)
    ax.tick_params(axis='x', labelsize=8)
    ax.set_yticklabels(sorted_features, fontsize=8)
    ax.set_xlim([0, 0.5])

    # Adjust the layout of the subplot
    fig.tight_layout()

    # Show the plot
    plt.show()

re_importances = dt.feature_importances_
show_feature_importances(re_importances)
```

Le modèle nous met à disposition les features les plus importantes pour la classification.

Cela est déjà très intéressant et nous permet de voir quelles sont les features qui ont le plus d'impact sur la classification.

Nous avons donc `histogram_mean` et `mean_value_of_short_term_variability` qui sont de loin les plus importantes selon le modèle.

Nous allons maintenant utiliser SHAP pour expliquer les prédictions du modèle.

```python
import shap

explainer = shap.TreeExplainer(dt, feature_names=X_test.columns.tolist())
shap_values = explainer.shap_values(X_test)

pd.DataFrame(shap_values, columns=X_test.columns.tolist()).head()
```

```python
import shap.plots

shap_values_plot = explainer(X_test)
shap.plots.bar(shap_values_plot)
```

```python
shap.plots.beeswarm(shap_values_plot)
```

Nous pouvons voir que les features les plus importantes sont `histogram_mean` et `mean_value_of_short_term_variability`, ce qui est cohérent avec ce que nous avons vu précédemment. Aussi, les autres features importantes données sont similaires à celles données par le modèle.


Grâce au graphique beeswarm, nous pouvons voir l'impacte des features de chaque observation sur la prédiction du modèle.

### Explication locale

Ici, nous allons analyser deux prédictions labelisées `pathological`. Une classifiée correctement et une classifiée incorrectement comme `normal`.

Le but est de comprendre pourquoi le modèle fait juste ou faux et ainsi de pouvoir l'améliorer en comprenant les features qui ont eu un impact sur la prédiction.

```python
correct_index = 0 # Index of a correct prediction
incorrect_index = 0 # Index of an incorrect prediction

for i in range(len(y_test)):
    if y_test.iloc[i] == 3 and y_pred_dt[i] == 3:
        correct_index = i
    if y_test.iloc[i] == 3 and y_pred_dt[i] == 1:
        incorrect_index = i
    if correct_index != 0 and incorrect_index != 0:
        break

correct_row = X_test.iloc[correct_index]
incorrect_row = X_test.iloc[incorrect_index]

comparison_df = pd.DataFrame({
    f"Correct (index {correct_index})": correct_row,
    f"Incorrect (index {incorrect_index})": incorrect_row
})

print("Index : ", correct_index, " | Valeurs réelles : ", y_test.iloc[correct_index], " | Prédictions : ", y_pred_dt[correct_index])
print("Index : ", incorrect_index, " | Valeurs réelles : ", y_test.iloc[incorrect_index], " | Prédictions : ", y_pred_dt[incorrect_index])

comparison_df.style.set_caption("Comparaison des valeurs de features")
```

```python
shap_values = shap_values = explainer(X_test)

print(f"Waterfall plot for correct prediction (index {correct_index}) true label: {y_test.iloc[correct_index]} predicted label: {y_pred_dt[correct_index]}")
shap.plots.waterfall(shap_values[correct_index], )

print(f"Waterfall plot for incorrect prediction (index {incorrect_index}) true label: {y_test.iloc[incorrect_index]} predicted label: {y_pred_dt[incorrect_index]}")
shap.plots.waterfall(shap_values[incorrect_index])
```

Grâce à SHAP, nous pouvons voir sur nos deux observations les features qui ont eu un impact sur la prédiction du modèle.

Ainsi on comprend les différences entre les deux observations et pourquoi l'une a été classifiée comme `pathological` et l'autre comme `normal`.
Premièrement, on remarque que les deux observations ont des valeurs de `histogram_mean` et `mean_value_of_short_term_variability` très différentes et qu'ils ont un impact très important sur la prédiction du modèle.
Ensuite, on remarque sur notre observation mal classifiée que plusieurs features (`histogram_number_of_peaks` et `histogram_median`) ont eu un impact négatif sur la prédiction du modèle, ce qui a conduit à une mauvaise classification.
Finalement, la feature `percentage_of_time_with_abnormal_long_term_variability` est à 0 sur l'observation mal classifiée alors qu'elle est à 86 sur l'observation bien classifiée, ce qui a également un impact important sur la prédiction du modèle.

Voyons maintenant comment LIME nous permet de faire la même chose.

```python
from lime.lime_tabular import *

explainer = LimeTabularExplainer(
    X_train.values,
    feature_names=X_train.columns.tolist(),
    class_names=['fetal_health'],
    mode='regression')
```

```python
#Show correct prediction
exp = explainer.explain_instance(
    data_row=X_test.iloc[correct_index], 
    predict_fn=dt.predict
)

exp.show_in_notebook(show_table=True)
```

```python
#Show incorrect prediction
exp = explainer.explain_instance(
    data_row=X_test.iloc[incorrect_index], 
    predict_fn=dt.predict
)

exp.show_in_notebook(show_table=True)
```

Nous pouvons voir que LIME nous donne globalement les mêmes features importantes que SHAP (`histogram_mean` et `percentage_of_time_with_abnormal_long_term_variability`), mais sous une forme différente. Cependant, l'ordre d'importance est différent.

### Résumé de l'analyse

Nous avons pu voir que les deux méthodes d'explication post-hoc, LIME et SHAP, nous permettent de comprendre les prédictions du modèle et de voir quelles sont les features importantes pour la classification.
Dans notre cas, les deux méthodes nous donnent des résultats assez similaires et nous permettent de comprendre les différences entre les observations classifiées correctement et incorrectement.

Nous avons pu voir que les features `histogram_mean`, `mean_value_of_short_term_variability` et `percentage_of_time_with_abnormal_long_term_variability` sont très importantes pour la classification et que les autres features ont un impact moins important mais peuvent tout de même influencer la prédiction du modèle.
